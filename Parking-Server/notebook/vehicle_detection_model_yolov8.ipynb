```python
# Validate the trained YOLOv8 model
metrics = model.val(data="/content/Car-Detection-Model-17/data.yaml")

# Print available metric keys (for debugging)
print("ðŸ”¹ Available Keys in metrics.results_dict:")
print(metrics.results_dict.keys())

# Readable formatted output
print("\nðŸ” **YOLOv8 Model Evaluation Metrics:**\n")

# Ensure the correct key names are used
map50 = metrics.results_dict.get('metrics/mAP50(B)', 'N/A')
map50_95 = metrics.results_dict.get('metrics/mAP50-95(B)', 'N/A')
precision = metrics.results_dict.get('metrics/precision(B)', 'N/A')
recall = metrics.results_dict.get('metrics/recall(B)', 'N/A')

print(f"ðŸ“Œ **Mean Average Precision (mAP@50):** {map50}")
print(f"ðŸ“Œ **Mean Average Precision (mAP@50-95):** {map50_95}")
print(f"ðŸ“Œ **Precision:** {precision}")
print(f"ðŸ“Œ **Recall:** {recall}")

# Compute F1 Score if precision and recall are available
if precision != 'N/A' and recall != 'N/A':
    f1_score = 2 * (precision * recall) / (precision + recall + 1e-6)
    print(f"ðŸ“Œ **F1 Score:** {f1_score:.4f}")
else:
    print("âš ï¸ **F1 Score cannot be computed due to missing values.**")

# Per-Class Performance
print("\nðŸš€ **Per-Class Performance:**")
for class_id, class_name in metrics.names.items():
    class_map = metrics.results_dict.get(f'classes/{class_id}/mAP50(B)', 'N/A')
    print(f"   âœ… {class_name}: mAP@50 = {class_map}")

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set theme
sns.set_style("whitegrid")

# Data
metrics_data = {
    "Precision": metrics.results_dict["metrics/precision(B)"],
    "Recall": metrics.results_dict["metrics/recall(B)"],
    "F1 Score": 2 * (
        metrics.results_dict["metrics/precision(B)"]
        * metrics.results_dict["metrics/recall(B)"]
    ) / (
        metrics.results_dict["metrics/precision(B)"]
        + metrics.results_dict["metrics/recall(B)"]
        + 1e-6
    ),
}

# Plot
plt.figure(figsize=(8, 6))
sns.barplot(x=list(metrics_data.keys()), y=list(metrics_data.values()), palette="coolwarm")
plt.ylim(0, 1)
plt.title("Model Performance Metrics")
plt.ylabel("Score")
plt.show()
```

```python
# Data
map_scores = {
    "mAP@50": metrics.results_dict["metrics/mAP50(B)"],
    "mAP@50-95": metrics.results_dict["metrics/mAP50-95(B)"],
}

# Plot
plt.figure(figsize=(8, 6))
sns.lineplot(x=list(map_scores.keys()), y=list(map_scores.values()), marker='o', linestyle='-', color='blue')
plt.ylim(0, 1)
plt.title("Mean Average Precision (mAP)")
plt.ylabel("Score")
plt.grid(True)
plt.show()
```
